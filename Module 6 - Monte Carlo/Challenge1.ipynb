* #### La Gaussienne 

La distribution gaussienne "built-in" est la suivante :

```python
x=numpy.random.normal(size=100000)
binwidth=0.10
pyplot.xlim(0,5)
bins=numpy.arange(0.,5.1, binwidth)
pyplot.hist(x,bins,normed=True);

xc = bins[:-1] + binwidth/2
pyplot.plot(xc,2*numpy.exp(-xc**2/2.0)/numpy.sqrt(2*numpy.pi), lw=3)
```

La méthode de rejection Von Neumann consiste a généré un nombre aleatoire g suivant une certaine distribution $p_1(x)$ (exponentielle dans notre cas), et a partir de celle ci, en extraire la distribution désirée $p_2(x)$ (dans notres cas la distribution gaussienne). Pour que cela fonctionne, $p_2(x)$ doit satisfaire a la condition :

$$p_2(x)=Kp_1(x)$$

pour une certaine constante K. 
Dans l'algorithme on procede de la maniere suivante : on tire un nombre aleatoire suivant une distribution uniforme entre 0 et 1, et on le "conserve" seulemment si il satisfait :
$$U<\frac{p_2(g)}{Kp_1(g)}$$

avec g un nombre aletoire suivant une distribution exponentielle. 

```python
# 100000 nombres aléatoire eront généré pour un K=1,4
N = 100000
K = 1.4

# Liste pour recevoir l'ensemble des nombres aleatoires g 
h=[]

for i in range(N):
    u1 = numpy.random.random_sample()
    
    # g est distribuer selon une distribution exponentielle
    g = numpy.log(1/(1-u1))
    
    # Deuxieme nombre aleatoire suivant une distribution gaussienne 
    u2 = numpy.random.random_sample()
    
    # Acceptance/rejection step : ratio=(p_1(g)/K*p_2(g))
    ratio = 2*numpy.exp(-g**2/2.0)/numpy.sqrt(2*numpy.pi) / \
                (K*numpy.exp(-g))
    if (u2<ratio):
        # Affiliation de g au set h
        h.append(g)

# Taille du set généré par la methode
print("Size of accepted set h and expected size 1/K:", len(h), N/K)

# Histogramme
binwidth=0.1
bins=numpy.arange(0.,5.1, binwidth)
pyplot.hist(h,bins,normed=True,color='blue');

# Centre des bins et graphique de la distribution attendue
xc = bins[:-1] + binwidth/2
pyplot.plot(xc,2*numpy.exp(-xc**2/2.0)/numpy.sqrt(2*numpy.pi),lw=3,color='green')
```

Comparaison des graphiques pour les deux manières de générer la distribution gaussienne :


```python
#Histogramme fonction python
binwidth=0.10
pyplot.xlim(0,5)
bins=numpy.arange(0.,5.1, binwidth)
pyplot.hist(x,bins,normed=True,color='blue');

#Histogramme Von Neumann
binwidth=0.1
bins=numpy.arange(0.,5.1, binwidth)
pyplot.hist(h,bins,normed=True,color='green');

#Graphe
xc = bins[:-1] + binwidth/2
pyplot.plot(xc,2*numpy.exp(-xc**2/2.0)/numpy.sqrt(2*numpy.pi),lw=3,color='red')
```

On remarque que les deux distributions sont extremmements semblables, on ne voit même pas le contraste entre l'histogramme vert et bleu ! Conclusions : la methode de réjection est une bonne methode pour générer un distribution gaussienne !

La méthode de distribution inverse consiste a généré un nombre aléatoire x suivant une certaine distribution $F(x)$ (distribution exponentielle dans notre cas) a partir de sa distribution inverse $F^{-1}(x)$.
En particulier, si $z$ suit une distribution exponnentielle te que  

$$z=1-e^{-\alpha y}$$

Alors on a : 

$$e^{-\alpha y}=1-z$$

et donc

$$y=-\frac{1}{\alpha}ln(1-z)$$

ou y est la distribution inverse de la distriution exponentielle. On peut donc généré un nombre aleatoire suivant une distribution exponentielle a partir de ce nombre aleatoire y suivant la distribution inverse :

```python
alpha = 0.25 # paramètre de la distribution exponentielle
z = numpy.random.random_sample(100000) 
y = -1/alpha*numpy.log(1-z) # Distribution inverse 

# Histogramme de la distribution inverse 
binwidth=0.3
pyplot.xlim(0,20)
bins=numpy.arange(0.,21., binwidth)
pyplot.hist(y,bins,normed=True);

# Definition du centre des bins et graphique de la distribution désirée
zc = bins[:-1] + binwidth/2
pyplot.plot(zc,alpha*numpy.exp(-alpha*zc), lw=2)
```

```python
#Histogramme fonction python
binwidth=0.3
pyplot.xlim(0,20)
bins=numpy.arange(0.,21., binwidth)
pyplot.hist(x,bins,normed=True,color="blue");

#Histogramme distribution inverse
binwidth=0.3
pyplot.xlim(0,20)
bins=numpy.arange(0.,21., binwidth)
pyplot.hist(y,bins,normed=True,color='green');

#Graphe
zc = bins[:-1] + binwidth/2
pyplot.plot(zc,alpha*numpy.exp(-alpha*zc), lw=2,color="red")
```

### 2) Methode de Box-Muller :

La méthode de Box-Muller sert a généré des variables aleatoires suivant une distribution gaussienne a partir de variables aleatoires suivant une distribution uniforme. Le principe est le suivant : on demarre par deux variables aleatoires $u_1$ et $u_2$ suivant une distribution uniforme entre 0 et 1.
Ensuite , en définissant deux nouvelle variable aléatoire comme : 
$$z_1=\sqrt{-2ln{(u_1)}}\cos{(2\pi u_2)}$$
et
$$z_2=\sqrt{-2ln{(u_1)}}\sin{(2\pi u_2)}$$
on crée deux variables $z_1$ et $z_2$ suivant une distribution Gaussienne !
On utilise ici la methode en cordonnées cartesiennes, elle peut aussi bien etre faite en coordonées polaire.

* #### Le code 

```python
# Changemment de variables
def gaussian(u1,u2):
  z1 = numpy.sqrt(-2*numpy.log(u1))*numpy.cos(2*numpy.pi*u2)
  z2 = numpy.sqrt(-2*numpy.log(u1))*numpy.sin(2*numpy.pi*u2)
  return z1,z2

# Deux distribution uniforme entre 0 et 1
u1 = numpy.random.rand(1000)
u2 = numpy.random.rand(1000)

# C'est parti
z1,z2 = gaussian(u1,u2)
```

Histogramme de z1 : 

```python
binwidth=0.3
pyplot.xlim(-5,5)
bins=numpy.arange(-21.,21., binwidth)
pyplot.hist(z1,bins,normed=True,color="blue");
```

Histogramme de z2 :

```python
binwidth=0.3
pyplot.xlim(-5,5)
bins=numpy.arange(-21.,21., binwidth)
pyplot.hist(z2,bins,normed=True,color="blue");
```
